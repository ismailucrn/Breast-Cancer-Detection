```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')   
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1, 'B':0})
```

```python
data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True) # drop id columns because this not useful for classification, and drop Unnamed: 32 because it is empty
```

```python
sns.countplot(data['diagnosis'], hue=data['diagnosis'])
```

```python
data_col = data.columns 
```

```python
for i in data_col:
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.histplot(data = data, x = i, hue = 'diagnosis', kde=True)
    plt.title(i)
    plt.show()
```

```python
corr = data.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
#corrention between diagnosis and other features
corr['diagnosis'].sort_values(ascending=False)
```
